- Cho

    2a + b + c = 4
    
    a + 3b + 2c = 5
    
    a = 6
 - Quy về Ax = B. Sau đó giải tìm x
    
    

```python
import numpy as np
```

```python
A = np.array([[2,1,1], [1,3,2], [1,0,0]])
```

```python
B = np.array([4,5,6])
```

```python
print("Solution [a, b, c]: \n", np.linalg.solve(A, B))
```

# Câu 2:
- Tạo 1 array points có 1000 phần tử cách đều nhau với start = -5, end = 5, step= 0.01
- Tạo ma trận điểm có 1000 điểm với x,y = points,point bằng np.meshgird
- Tính z = $\sqrt{x^2 + y^2}$ thông qua ma trận điểm
- Biểu diễn z


```python
points = np.arange(-5, 5, 0.01)
points[:5]
```

```python
points.size
```

```python
x, y = np.meshgrid(points, points)
x.size
```

```python
y.size
```

```python
z = np.sqrt(x**2 + y**2)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (8,8))
plt.imshow(z, cmap = plt.cm.Blues)
plt.colorbar()
plt.title("Image plot of $\sqrt{x^2 + y^2}$ for a grid of values")
plt.show()
```
